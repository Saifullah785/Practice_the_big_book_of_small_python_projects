```python
import sys


bitmap = """
....................................................................
   **************   *  *** **  *      ******************************
  ********************* ** ** *  * ****************************** *
 **      *****************       ******************************
          *************          **  * **** ** ************** *
           *********            *******   **************** * *
            ********           ***************************  *
   *        * **** ***         *************** ******  ** *
               ****  *         ***************   *** ***  *
                 ******         *************    **   **  *
                 ********        *************    *  ** ***
                   ********         ********          * *** ****
                   *********         ******  *        **** ** * **
                   *********         ****** * *           *** *   *
                     ******          ***** **             *****   *
                     *****            **** *            ********
                    *****             ****              *********
                    ****              **                 *******   *
                    ***                                       *    *
                    **     *                    *
...................................................................."""

print('Enter the message to display with the bitmap.')
message = input('> ')
if message == '':
  sys.exist()

# Loop over each line in the bitmap:
for line in bitmap.splitlines():
  for i, bit in enumerate(line):
    if bit == ' ':
      print('',end='')
    else:
      print(message[i % len(message)], end='')
  print()
```
